## How to find the row in a Pandas DataFrame with the maximum value for a column in Python
Finding the row in a Pandas DataFrame with the maximum value for a column results in the the row which has the largest element in a specified column.

- USE pd.Series.idxmax() TO FIND THE ROW WITH THE MAXIMUM VALUE FOR A COLUMN
- Call pd.Series.idxmax() with pd.Series as a column in a DataFrame to find the index of the maximum value in that column. 
- Then, use pd.DataFrame.iloc to extract the row with that position.

```python
max_index = ass["Data.New HIV Infections.All Ages"].idxmax()
```

```python
max_row = ass.iloc[[max_index]]
```

```python
print(max_row)
```

From the list printed aove, it appears the country with the highest number of reported Aids cases is **South Africa**
